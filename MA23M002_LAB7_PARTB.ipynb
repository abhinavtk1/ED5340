5. WAP to plot a 3-d graph of the helical wave signal using the scatter method and normal line method. Plot them separately and specify legend.

```python
import numpy as np
import matplotlib.pyplot as plt

def helical_wave_signal(t):
    x = np.cos(t)
    y = np.sin(t)
    return x, y

# Generate data points
t = np.linspace(0, 25, 1000)
x, y = helical_wave_signal(t)
```

```python
# Scatter plot
fig1 = plt.figure(figsize=(10, 6))
ax1 = fig1.add_subplot(121, projection='3d')
ax1.scatter(x, y, t, c='r', label='Scatter Method')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('t')
ax1.legend()
# Normal line method
ax2 = fig1.add_subplot(122, projection='3d')
ax2.plot(x, y, t, c='b', label='Normal Line Method')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('t')
ax2.legend()
plt.suptitle('Helical Wave Signal')
plt.tight_layout()
plt.show()
```

8.Take a N X M integer array matrix with space separated elements ( N = rows and M  = columns). Your task is to print the transpose and flatten results using numpy

```python
import numpy as np

#  Input array of shape 3x4 as string type
inp_arr = []
N = int(input("Enter no. of rows: "))
M = int(input("Enter no. of cols: "))
for _ in range(N):
  row = input("Enter items of each row as space separated elements: ")
  inp_arr.append(row)

# Convert inp_arr to a numpy array
arr = np.array([[int(x) for x in item.split()] for item in inp_arr])

# Transpose the array using numpy
arr_T = arr.T

# Flatten the array
arr_flatten = arr.flatten()

# Print inputs and results
print("Input Array:")
print(arr)
print("\nTranspose Array:")
print(arr_T)
print("\nFlattened Array:")
print(arr_flatten)
```

10. Use the central difference method to find the first and second order derivatives of the function. Use the following function for testing the result. And also verify the result manually (Write on paper and upload jpg). <br>
Refer to section 2.5.1 of “Optimization for Engineering Design: Algorithms and Examples” by KALYANMOY DEB, 2nd edition

f(x) = 3x**2 + 2x



```python
def first_derivative(f, x, delx):
  return (f(x + delx) - f(x - delx)) / (2 * delx)

def second_derivative(f, x, delx):
  return (f(x + delx) - 2*f(x) + f(x - delx)) / (delx**2)

def f(x):
  return 3*(x**2) + 2*x

x = 1         # Test the functions at x = 1
delx = 0.01

f_dash = first_derivative(f, x, delx)             # Calculate first order derivative of function at x=1
f_double_dash = second_derivative(f, x, delx)     # Calculate first order derivative of function at x=1

print("First order derivative at x =", x, "is:", f_dash)
print("Second order derivative at x =", x, "is:", f_double_dash)
```
