6. Create a base class Vehicle with the following attributes: <br>
make (string) <br>
model (string) <br>
year (int) <br>
Create a method initialize_vehicle to set the above attributes. Also, create a method display_vehicle to print these attributes. <br>
Create a class Car inherited from Vehicle with the following additional attribute:<br>
fuel_type (string) <br>
Create a method get_car_details to initialize the above attribute along with Vehicle attributes.<br>
Also, create a method display_vehicle to print these attributes along with Vehicle attributes. <br>
Create a class Bike inherited from Vehicle with the following additional attribute:<br>
gear_count (int) <br>
Create a method get_bike_details to initialize the above attribute along with Vehicle attributes. <br>
Also, create a method display_vehicle to print these attributes along with Vehicle attributes.<br>
Create two different objects for Car and Bike and demonstrate each of the methods.

```python
class Vehicle:
    def __init__(self):
      self.make = ""
      self.model = ""
      self.year = 0

    def initialize_vehicle(self, make, model, year):
      self.make = make
      self.model = model
      self.year = year

    def display_vehicle(self):
      print("Make: ", self.make, ", Model: ", self.model, ", Year: ", self.year)


class Car(Vehicle):
    def __init__(self):
      super().__init__()
      self.fuel_type = ""

    def get_car_details(self, make, model, year, fuel_type):
      self.initialize_vehicle(make, model, year)
      self.fuel_type = fuel_type

    def display_vehicle(self):
      print("Printing car details:")
      super().display_vehicle()
      print("Fuel Type:", self.fuel_type)


class Bike(Vehicle):
    def __init__(self):
      super().__init__()
      self.gear_count = 0

    def get_bike_details(self, make, model, year, gear_count):
      self.initialize_vehicle(make, model, year)
      self.gear_count = gear_count

    def display_vehicle(self):
      print("Printing bike details:")
      super().display_vehicle()
      print("Gear Count:", self.gear_count)
```

```python
carObj = Car()
carObj.get_car_details("Toyota", "Camry", 2020, "Petrol")
carObj.display_vehicle()
```

```python
bikeObj = Bike()
bikeObj.get_bike_details("Yamaha", "YZF R1", 2021, 6)
bikeObj.display_vehicle()
```

7. Suppose you are building a Python program to manage a school's student data. You need to create a Student class that contains information such as the student's name, age, grade, and class schedule. How can you use class variables in Python to define these shared attributes of the Student class? What are the advantages of using class variables in this scenario? Can you provide an example program that demonstrates the use of class variables in the Student class?

```python
class Student:
  # Class variables
  school_name = "DAV public school"
  total_students = 50
  classes_offered = 4


  def __init__(self, name, age, grade, schedule):
    self.name = name
    self.age = age
    self.grade = grade
    self.schedule = schedule

  def student_info(self):
    print("Student details: ")
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Grade: {self.grade}")
    print("Class Schedule in order :", *self.schedule)
    print()
```

# Advantages of class variables:
1. They promote memory efficiency by being shared among all instances of a class, thus saving memory by avoiding the creation of multiple copies of the same data.
2. Facilitate shared state across all instances of a class, enabling all instances to access and modify the same data easily
3. Class variables are easy to access using the class name itself without requiring an instance of the class, enhancing code readability and accessibility

8. Class Inheritance in Python: Finding GCD (greatest common divisor) and LCM (least common multiple) of Numbers and Handling Composite Numbers.

a) Create a Numbers class with a, b, find_gcd(), and find_lcm() methods.

b) Create an EvenNumbers class that inherits from Numbers and overrides find_lcm() to handle even numbers.

c) Create an OddNumbers class that inherits from Numbers and overrides find_lcm() to handle odd numbers.

d) Create a CompositeNumbers class that inherits from EvenNumbers and OddNumbers and overrides find_gcd() to handle composite numbers.

e) Create a CompositeNumbers object with a = 12 and b = 9, and call its find_lcm() and find_gcd() methods.

```python
class Numbers:
  def __init__(self, a, b):
    self.a = a
    self.b = b

  def find_gcd(self):
    x = self.a
    y = self.b
    if x > y:
        smallest = y
    else:
        smallest = x
    for i in range(smallest, 0, -1):
        if((x % i == 0) and (y % i == 0)):
            gcd = i
            break
    return gcd

  def find_lcm(self):
    gcd = self.find_gcd()
    lcm = (self.a * self.b) // gcd
    return lcm
```

```python
class EvenNumbers(Numbers):
  def find_lcm(self):
    if self.a % 2 == 0 and self.b % 2 == 0:   # check for even
        return super().find_lcm()
    else:
        print("Both no.s are not even")
        return None

class OddNumbers(Numbers):
    def find_lcm(self):
        if self.a % 2 != 0 and self.b % 2 != 0:
            return super().find_lcm()
        else:
            print("Both no.s are not odd")
            return None

class CompositeNumbers(EvenNumbers, OddNumbers):
      def check_composite(self,n):
        if n <= 1:
          return False
        for i in range(2, int(n**0.5) + 1):
          if n % i == 0:
              return True
        return False

      def find_gcd(self):
        if self.check_composite(self.a) and self.check_composite(self.b):
          return super().find_gcd()
        else:
          print("Both are not composite")
          return None

      def find_lcm(self):
        return (self.a * self.b) // super().find_gcd()
```

```python
# Create a CompositeNumbers object with a = 12 and b = 9
composite_obj = CompositeNumbers(12, 9)

# Call its find_lcm() and find_gcd() methods
print("GCD:", composite_obj.find_gcd())
print("LCM:", composite_obj.find_lcm())
```

9. WAP to manage the collections of books in a library in the following  manner:

 Create a Python script that can both read from and write to a CSV file, containing details about each book. Each book's information will include its title, author, publication year, and ISBN number. Your script should be capable of adding new books to the CSV file and listing all the books currently stored in the file.

The program should begin by checking if the CSV file exists. If it does not, your script should create it and initialize it with the appropriate headers. Then, there should be 2 options: to add a new book or to display all books. When adding a new book, the user should be prompted to enter the title, author, publication year, and ISBN number. This new book should then be added to the CSV file without overwriting the existing entries. When choosing to display all books, the script should read from the CSV file and print each book's details.



```python
import csv
import os

# Function to create the CSV file
def create_csv(filename):
  with open(filename, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Author", "Publication Year", "ISBN"])

# Add a book to the CSV file
def add_book(filename, title, author, year, isbn):
    with open(filename, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([title, author, year, isbn])
    print("Book added to database ")
    print()

# Function to show all books from the CSV file
def display_books(filename):
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            print("Title:", row[0])
            print("Author:", row[1])
            print("Publication Year:", row[2])
            print("ISBN:", row[3])
            print("\n")
```

```python
csv_filename = "book_list.csv"
if not os.path.exists(csv_filename):
  create_csv(csv_filename)    # Check if the CSV file exists, otherwise create it

# Add a new book by taking input from user
title = input("Enter the title: ")
author = input("Enter the author: ")
year = input("Enter the publication year: ")
isbn = input("Enter the ISBN: ")

add_book(csv_filename, title, author, year, isbn)

# Display books
display_books(csv_filename)
```

```python
csv_filename = "booklist1.csv"    # csv file doesn't exist
if not os.path.exists(csv_filename):
  create_csv(csv_filename)    # Check if the CSV file exists, otherwise create it

# Add a new book
title = input("Enter the title: ")
author = input("Enter the author: ")
year = input("Enter the publication year: ")
isbn = input("Enter the ISBN: ")

add_book(csv_filename, title, author, year, isbn)

# Display books
display_books(csv_filename)
```

10. WAP to create a pandas dataframe with a list of words and sort them in ascending order. The sorted words should be copied to a new file.
