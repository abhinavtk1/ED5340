1. Write a program that takes coefficients A, B, C, D, and E as inputs representing a 4th degree polynomial in the form Ax^4 + Bx^3 + Cx^2 + Dx + E. Calculate the values of this polynomial for x in the range from -100 to 100, with constant discrete intervals. <br>
Store the resulting x and y values as a NumPy array, where x represents the input values, and y represents the corresponding output values of the polynomial. Finally, use Matplotlib to plot the graph using the generated NumPy array.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

def polynomialfn(x, A, B, C, D, E):
  return A*x**4 + B*x**3 + C*x**2 + D*x + E

x = np.arange(-100, 101, 5)
A = float(input("Enter coefficient A: "))
B = float(input("Enter coefficient B: "))
C = float(input("Enter coefficient C: "))
D = float(input("Enter coefficient D: "))
E = float(input("Enter coefficient E: "))
y = polynomialfn(x, A, B, C, D, E)
```

```python
# Plotting using matplotlib
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot y vs x')
plt.grid(True)
plt.show()
```

2. Suppose you have a dictionary containing information about monthly sales for different products over a period of time. The dictionary has the following structure.

sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

Write a Python script to convert this dictionary into a pandas DataFrame, calculate the total sales for each product over the entire period, and then create a bar plot using matplotlib to visualize the total sales for each product.

```python
sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]
    }
```

```python
import pandas as pd
df = pd.DataFrame(sales_data)
```

```python
df
```

```python
# calculate the total sales for each product over the entire period
total_sales = df.groupby('Product')['Sales'].sum().reset_index()
print(total_sales)
```

```python
# Plot the total sales for each product using bar plot
plt.figure(figsize=(4, 3))
plt.bar(total_sales['Product'], total_sales['Sales'], color='blue', width = 0.5)
plt.title('Total Sales for Each Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

Generate surface plots for these multi-variable functions over the range

x=[−10,10] and y=[−10,10] , ensuring to add a title and labels for all axes:

(1) z = cos(sqrt(x^2+y^2)

(2) z = e^(-(x^2+y^2))

```python
# Define x and y
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
```

(1) z = cos(sqrt(x^2+y^2)

```python
# Define the functions
z = np.cos(np.sqrt(x**2 + y**2))

# Surface plot
fig1 = plt.figure(figsize=(8, 6))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(x, y, z)
ax1.set_title('z = cos(sqrt(x^2+y^2))')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
plt.show()
```

(2) z = e^(-(x^2+y^2))



```python
# Define the function
z = np.exp(-(x**2 + y**2))

# Surface plot
fig1 = plt.figure(figsize=(8, 6))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(x, y, z)
ax1.set_title('Function value Z')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
plt.show()
```

(3) z =  log(x^2+y^2) where x^2+y^2>0




```python
z = np.log(x**2 + y**2)
z[x**2 + y**2 <= 0] = np.nan
# Surface plot
fig1 = plt.figure(figsize=(8, 6))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(x, y, z)
ax1.set_title('Function value Z')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
plt.show()
```

4. For the function J(w) = w^2 + (54/w), implement the bracketing method (choose your own a, b, n).



```python
def J(w):
    return w**2 + 54/w

def bracketing_method(a, b, n):
    tol = 0.01        # tolerance to terminate if reached near minimum
    dw = (b - a)/n
    i = dw
    w1 = a
    w2 = w1 + dw
    w3 = w2 + dw
    while i < n:
      if J(w1)>= J(w2) and J(w3)>=J(w2):
        if abs(w1-w2)<= tol:
          return w1
        bracketing_method(w1, w3, n)
      else:
        w1 = w2
        w2 = w3
        w3 = w2 + dw
        if w3 <= b:
          continue
        else:
          print("No min exists between given pair of end points")
          return None
```

```python
# Checking fot different intervals
w = bracketing_method(2,4,1000)
if w:
  print("The min value is at w = : ", w)
  print("Min value is: ", J(w))
```
